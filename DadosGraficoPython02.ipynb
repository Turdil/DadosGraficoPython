```python
import pandas as pd
from google.colab import files
from IPython.display import display
import matplotlib.pyplot as plt

colunas=["Nome","Idade","Preferência(Doce/Salgado)","Cor Favorita","Horas de Sono","Gênero Musical"]

tb=pd.DataFrame(columns=colunas)
display(tb)

arquivo_excel="dados_modelos.xlsx"

tb.to_excel(arquivo_excel,index=False,engine="openpyxl")

files.download(arquivo_excel)
print("Baixe o arquivo e preencha corretament com os dados coletados")
```

```python
subir = files.upload()
nome_arquivo = list(subir.keys())[0]
print(f'Arquivo{nome_arquivo}')
```

```python
dados=pd.read_excel(nome_arquivo,engine="openpyxl")
display(dados)
```

```python
print("Cabeçalho")
display(dados.head())
```

```python
print(f'Media de Idade: {dados["Idade"].mean():.2f} anos')
print(f'Media de Horas de sono: {dados["Horas_sono"].mean():.2f} h')
print("\n Pref. Alimentar")
print(dados["Comida_preferida"].value_counts())
dados['Genero_musical'].value_counts().plot(kind='bar',title="Musical")
plt.xlabel("Gênero")
plt.ylabel("Numero de alunos")
plt.grid(True)
plt.show()
```

```python
dados['Cor_preferida'].value_counts().plot(kind='line',title="Cor_preferida")
plt.xlabel("Cor")
plt.ylabel("Numero de alunos")
plt.grid(True)
plt.show()
```

```python
dados_removidos = dados.drop_duplicates().reset_index(drop=True)
display(dados_removidos)
```
